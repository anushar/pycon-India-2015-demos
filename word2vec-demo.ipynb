#### Play around with pretrained word2vec models

```python
from  gensim.models.word2vec import Word2Vec
```

```python
model = '../GoogleNews-vectors-negative300.bin'
word2vec_model = Word2Vec.load_word2vec_format(model, binary=True)
```

```python
word2vec_model.most_similar(positive=['woman', 'king'], negative=['man'])
```

```python
word2vec_model.doesnt_match("breakfast cereal dinner lunch".split())
```

```python
word2vec_model.similarity('woman', 'man')
```

```python
word2vec_model['computer']  # raw numpy vector of a word
```

```python
word2vec_model.n_similarity(['sushi', 'shop'], ['japanese', 'restaurant'])
```

```python
word2vec_model.most_similar('modi')
```

```python
# is a word is not there it will throw a keyError
word2vec_model.most_similar('pycon')
```
